# Making Unique 3-Image Combinations [target, distractor 1, distractor 2]

```python
import os
```

```python
path = '.'
files = os.listdir(path)
fileDir = []
for file in files:
    if ".png" in file:
        fileDir.append(file)
```

```python
fileDir = fileDir[:-8]
```

```python
import itertools
import random
```

```python
unique_comb3 = set(itertools.permutations(fileDir, 3))
random.shuffle(list(unique_comb3))
unique_comb3= list(unique_comb3)[:200]
```

```python
unique_3_list = [list(t) for t in unique_comb3]
for i in range(len(unique_comb3)):
    for j in range(len(unique_3_list[0])):
        unique_3_list[i][j] = f'images/{unique_3_list[i][j]}'

unique_comb3 = [tuple(l) for l in unique_3_list]
unique_comb3
```

```python
import csv
```

```python
with open("output3.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(unique_comb3)
```
